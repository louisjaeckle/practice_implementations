```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv('concrete.csv')
df.head()
```

```python
arr = df.to_numpy()
arr = arr[:,5:7].tolist()
```

```python
def get_cluster_points(cluster_means,data):
    cluster_points = [[] for _ in range(len(cluster_means))]
    for point in data:
        min_dist = -1
        cluster_ind = 0
        for clust in range(len(cluster_means)):
            dist = math.sqrt(sum([(cluster_means[clust][i]-point[i])**2 for i in range(2)]))
            if min_dist==-1 or dist<min_dist:
                min_dist = dist
                cluster_ind = clust
        cluster_points[cluster_ind].append(point)
    return cluster_points

def reset_cluster_means(cluster_means, data):
    cluster_points = get_cluster_points(cluster_means,data)
    for clust in range(len(cluster_means)):
        if len(cluster_points[clust])==0:
            continue
        mean_x = sum([point[0] for point in cluster_points[clust]])/len(cluster_points[clust])
        mean_y = sum([point[1] for point in cluster_points[clust]])/len(cluster_points[clust])
        cluster_means[clust] = [mean_x,mean_y]
    return cluster_means
```

```python
plt.scatter([point[0] for point in arr],[point[1] for point in arr])
plt.show()
```

```python
arr_xs = [point[0] for point in arr]
arr_ys = [point[1] for point in arr]

x_min = min(arr_xs)
x_max = max(arr_xs)
y_min = min(arr_ys)
y_max = max(arr_ys)
```

```python
K = 4
```

```python
cluster_xs = np.random.uniform(x_min,x_max,size=(K,))
cluster_ys = np.random.uniform(y_min,y_max,size=(K,))
cluster_means = [[cluster_xs[i],cluster_ys[i]] for i in range(K)]
cluster_means
```

```python
def plot_clusters(cluster_means,data):
    cluster_xs = [point[0] for point in cluster_means]
    cluster_ys = [point[1] for point in cluster_means]
    plt.scatter(cluster_xs,cluster_ys,color='k',s=100)
    
    cluster_points = get_cluster_points(cluster_means,data)
    colors = ['m','r','g','c']
    for ind in range(len(cluster_means)):
        xs = [point[0] for point in cluster_points[ind]]
        ys = [point[1] for point in cluster_points[ind]]
        plt.scatter(xs,ys,color = colors[ind])
    plt.show()

plot_clusters(cluster_means,arr)
```

```python
cluster_means = reset_cluster_means(cluster_means,arr)
plot_clusters(cluster_means,arr)
```

```python
def k_means_cluster(data,K, max_iter=100, plot=False):
    arr_xs = [point[0] for point in data]
    arr_ys = [point[1] for point in data]

    x_min = min(arr_xs)
    x_max = max(arr_xs)
    y_min = min(arr_ys)
    y_max = max(arr_ys)
    
    cluster_xs = np.random.uniform(x_min,x_max,size=(K,))
    cluster_ys = np.random.uniform(y_min,y_max,size=(K,))
    cluster_means = [[cluster_xs[i],cluster_ys[i]] for i in range(K)]
    
    for iter in range(max_iter):
        new_clusters = reset_cluster_means(cluster_means,arr)
        if new_clusters==cluster_means:
            break
        cluster_means = new_clusters
    if plot_clusters:
        plot_clusters(cluster_means,data)
    return get_cluster_points(cluster_means,data)
```

```python
a = k_means_cluster(arr,4,plot=True)
```
